```python
# Import python libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
```

```python
# Importing the dataset
dataset = pd.read_csv('Churn_modelling.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:13]
y = dataset.iloc[:, 13]
```

```python
y
```

```python
# create dummy variables
geography = pd.get_dummies(X['Geography'], drop_first=True)
gender = pd.get_dummies(X['Gender'], drop_first=True)
```

```python
geography
```

```python
gender
```

```python
# Concatenate the data frames
X = pd.concat([X, geography, gender], axis=1)
```

```python
# Drop unnecessary columns
X = X.drop(['Geography', 'Gender'], axis=1)
```

```python
X
```

```python
# Splitting the dataset into Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Feature scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train.shape
```

```python
# Initializing the ANN
classifier = tf.keras.models.Sequential()

# Adding the input layer and the first hidden layer
classifier.add(tf.keras.layers.Dense(units = 6, kernel_initializer = 'he_uniform', activation = 'relu', input_dim = 11))

# Add the second hidden layer
classifier.add(tf.keras.layers.Dense(units = 6, kernel_initializer = 'he_uniform', activation = 'relu'))

# Adding the output layer
classifier.add(tf.keras.layers.Dense(units = 1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the training set
model_history=classifier.fit(X_train, y_train,validation_split=0.33, batch_size = 10, nb_epoch = 100)
```

```python
# list all data in history
print(model_history.history.keys())
# Summarize history for keys
plt.plot(model_history.history['acc'])
plt.plot(model_history.history['val_acc'])
plt.title('model')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc = 'upper left')
plt.show()
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

# Calculate the Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
```
